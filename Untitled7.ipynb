```python
telco = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn (2).csv")
telco
```

```python
telco.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=telco, palette='Set2')
plt.title('Customer Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()
```

 In this graphy more than 5000 customers need this service and 1000 to 2000 customers didnot need this service

```python
plt.figure(figsize=(8,5))
sns.histplot(data=telco, x='MonthlyCharges', hue='Churn', kde=True, palette='Set1', bins=30)
plt.title('Monthly Charges Distribution by Churn')
plt.xlabel('Monthly Charges')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
contract_churn = telco.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
contract_churn['Yes'].plot(kind='bar', color='salmon')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.xlabel('Contract Type')
plt.ylim(0, 1)
plt.show()
```

in this graphy we can see the churn rate by 0.0 to 1 and in the contract type month to month higher in the churn rate (0.0 to 0.4 above)

```python
sns.countplot(data=telco,x='gender',hue='Churn',palette="Set1")
plt.title("churn by gender")
plt.show()
```

in the graphy there are count of male and female count . female graphy (500 to 1000) chance to leave and in male graphy there are also (500 to  1000) have a chance to leave .

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn','customerID'
]
```

```python
label_encoder = LabelEncoder()
```

```python
for col in categorical_columns:
    telco[col] = label_encoder.fit_transform(telco[col])
```

```python
print(telco.head())
```

```python
le = LabelEncoder()
telco["gender_e"] = le.fit_transform(telco["gender"])
telco["Partner_e"] = le.fit_transform(telco["Partner"])
telco["PhoneService_e"] = le.fit_transform(telco["PhoneService"])
telco["MultipleLines_e"] = le.fit_transform(telco["MultipleLines"])
telco["InternetService_e"] = le.fit_transform(telco["InternetService"])
telco["OnlineSecurity_e"] = le.fit_transform(telco["OnlineSecurity"])
telco["DeviceProtection_e"] = le.fit_transform(telco["DeviceProtection"])
telco["TechSupport_e"] = le.fit_transform(telco["TechSupport"])
telco["StreamingTV_e"] = le.fit_transform(telco["StreamingTV"])
telco["StreamingMovies_e"] = le.fit_transform(telco["StreamingMovies"])
telco["Contract_e"] = le.fit_transform(telco["Contract"])
telco["PaperlessBilling_e"] = le.fit_transform(telco["PaperlessBilling"])
telco["PaymentMethod_e"] = le.fit_transform(telco["PaymentMethod"])
telco["Churn_e"] = le.fit_transform(telco["Churn"])
telco["customerID_e"] = le.fit_transform(telco["customerID"])
telco
```
